```python
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
```

```python
train_data_path = "C:/Users/soham/Downloads/archive/tomato/train"
validation_data_path = "C:/Users/soham/Downloads/archive/tomato/val"
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1,5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip(images_arr,axes):
        ax.imshow(img)
    plt.tight_layout()
    plt.show()
```

```python
training_datagen = ImageDataGenerator(rescale=1./255,
                                     rotation_range=40,
                                     width_shift_range=0.2,
                                     height_shift_range=0.2,
                                     shear_range=0.2,
                                     zoom_range=0.2,
                                     horizontal_flip=True,
                                     fill_mode='nearest')

training_data = training_datagen.flow_from_directory(train_data_path,
                                                    target_size=(150,150),
                                                    batch_size=32,
                                                    class_mode="binary")
```

```python
training_data.class_indices
```

```python
valid_datagen = ImageDataGenerator(rescale=1./255)

valid_data = valid_datagen.flow_from_directory(validation_data_path,
                                                    target_size=(150,150),
                                                    batch_size=32,
                                                    class_mode="binary")
```

```python
valid_data.class_indices
```

```python
images = [training_data[0][0][0] for i in range(5)]
plotImages(images)
```

```python
model_path = "C:/Users/soham/Downloads/archive/tomato/tomato_disease_model.h5"
checkpoint = ModelCheckpoint(model_path, monitor='val_accuracy', verbose=1,save_best_only=True,mode='max')
callbacks_list = [checkpoint]
```

```python
cnn_model = keras.models.Sequential([
                                     keras.layers.Conv2D(filters=32,kernel_size=3,input_shape=[150,150,3]),
                                     keras.layers.MaxPooling2D(pool_size=(2,2)),
                                     keras.layers.Conv2D(filters=64, kernel_size=3),
                                     keras.layers.MaxPooling2D(pool_size=(2,2)),
                                     keras.layers.Conv2D(filters=128, kernel_size=3),
                                     keras.layers.MaxPooling2D(pool_size=(2,2)),
                                     keras.layers.Conv2D(filters=256, kernel_size=3),
                                     keras.layers.MaxPooling2D(pool_size=(2,2)),
                                
                                     keras.layers.Dropout(0.5),
                                     keras.layers.Flatten(),
                                     keras.layers.Dense(units=128, activation='relu'),
                                     keras.layers.Dropout(0.1),
                                     keras.layers.Dense(units=256, activation='relu'),
                                     keras.layers.Dropout(0.25),
                                     keras.layers.Dense(units=10, activation='softmax'),
])

cnn_model.compile(optimizer=Adam(lr=0.0001), loss='sparse_categorical_crossentropy',metrics = ['accuracy'])
```

```python
cnn_model.summary()
```

```python
history = cnn_model.fit(training_data, epochs=50,verbose=1,validation_data = valid_data, callbacks = callbacks_list)
```
